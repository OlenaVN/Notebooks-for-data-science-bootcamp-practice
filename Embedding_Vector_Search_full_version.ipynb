Create a function that determines the cosine similarity between a vector and a batch of vectors using the cosine distance formula and the numpy library.

```python
import numpy as np
from numpy.linalg import norm

def cos_sim_using_numpy(a,b):
  cosine = np.dot(a, b) / (norm(a, axis=1) * norm(b))
  return ("Cosine Similarity:", cosine)
```

```python
A = np.array([[2, 1, 2], [3, 2, 9], [-1, 2, -3]])
B = np.array([3, 4, 2])
cos_sim_using_numpy(A,B)
```

```python
def cos_similarity_detector(embed_list):
  pair_list = list(itertools.combinations(embed_list,2))
  sim_list = []
  for i in range(len(pair_list)):
     sim = util.cos_sim(pair_list[i][0],pair_list[i][1]).tolist()[0][0]
     sim_list.append(sim)
  sim_list.sort()
  return sim_list
```
